```python
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from radis import calc_spectrum
import numpy as np

# Parametry
temperatures = [1200, 1500, 1800, 2100]
wmin, wmax = 4283.5, 4286.0  # cm
n = len(temperatures)

# Barevn kly
colors_co = plt.cm.plasma(np.linspace(0.33, 0.99, n))
cmap_h2o = LinearSegmentedColormap.from_list("blue_to_turquoise", [(0, 0, 0.5), (0.2, 1, 1)])
colors_h2o = [cmap_h2o(i) for i in np.linspace(0, 1, n)]

# Inicializace grafu
fig, ax = plt.subplots(figsize=(10, 5))

# Uchovn linek pro legendu
lines_all, labels_all = [], []

# CO absorbance
for T, color in zip(temperatures, colors_co):
    s_co = calc_spectrum(wmin, wmax, molecule='CO', isotope='1', pressure=1.01325,
                         Tgas=T, mole_fraction=0.02, path_length=10, databank='hitemp')
    wn, absorb = s_co.get("absorbance", wunit="cm-1", return_units=False)
    wavelength_nm = 1e7 / wn
    line, = ax.plot(wavelength_nm, absorb, color=color, label=f"CO ({T} K)")
    lines_all.append(line)
    labels_all.append(f"CO ({T} K)")

# HO absorbance
for T, color in zip(temperatures, colors_h2o):
    s_h2o = calc_spectrum(wmin, wmax, molecule='H2O', isotope='1', pressure=1.01325,
                          Tgas=T, mole_fraction=0.2, path_length=10,
                          databank=r'C:\Users\venan\.radisdb\hitemp\h2o\01_4150-4500_HITEMP2010.par')
    wn, absorb = s_h2o.get("absorbance", wunit="cm-1", return_units=False)
    wavelength_nm = 1e7 / wn
    line, = ax.plot(wavelength_nm, absorb, color=color, linestyle='dashed', label=f"HO ({T} K)")
    lines_all.append(line)
    labels_all.append(f"HO ({T} K)")

# Popisy os
ax.set_xlabel("Wavelength [nm]", fontsize=14)
#ax.set_ylabel("Absorbance ($\\ln(I_0/I)$)", fontsize=14)
ax.set_ylabel("Absorbance (HITEMP)", fontsize=14)
ax.tick_params(labelsize=12)
ax.grid(True, linestyle='--', alpha=0.5)

# Horn osa s vlnovm slem
def wavelength_to_wavenumber(x):
    return 1e7 / x

def wavenumber_to_wavelength(x):
    return 1e7 / x

secax = ax.secondary_xaxis('top', functions=(wavelength_to_wavenumber, wavenumber_to_wavelength))
secax.set_xlabel("Wavenumber [cm]", fontsize=14, labelpad=10)
secax.tick_params(labelsize=12)

# Jedna spolen legenda
#legend_title = " HITEMP database: \n ($L_{opt}=10$ cm and $p=1$ atm) \n ($X_{CO} = 0.02$, $X_{H_2O} = 0.2$)"
ax.legend(lines_all, labels_all, loc='upper left', ncol=2, fontsize=10)#, title=legend_title, title_fontsize=11)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.colors import LinearSegmentedColormap
from radis import calc_spectrum
import numpy as np

# === Parametry ===
temperature = 1500
wmin, wmax = 4283.5, 4286.0  # cm
x_co = 0.02
x_h2o = 0.2

# === Barvy ===
color_co = plt.cm.plasma(0.55)
#color_co = (0.820, 0.235, 0.387)
#color_h2o = (0.2, 1, 1)  # tyrkysov
color_h2o = (51/255, 153/255, 204/255)

# === Inicializace grafu ===
fig, ax1 = plt.subplots(figsize=(10, 5))
ax2 = ax1.twinx()

# === HITEMP CO ===
s_co = calc_spectrum(wmin, wmax, molecule='CO', isotope='1', pressure=1.01325,
                     Tgas=temperature, mole_fraction=x_co, path_length=10, wstep=0.001,
                     databank='hitemp')
wn_co, absorb_co = s_co.get("absorbance", wunit="cm-1", return_units=False)
wavelength_co = 1e7 / wn_co
line_co, = ax1.plot(wavelength_co, absorb_co, color=color_co, linewidth=2.0, label=f"CO ({temperature} K)", zorder=2)

# === HITEMP HO ===
s_h2o = calc_spectrum(wmin, wmax, molecule='H2O', isotope='1', pressure=1.01325,
                      Tgas=temperature, mole_fraction=x_h2o, path_length=10, wstep=0.001,
                      databank=r'C:\Users\venan\.radisdb\hitemp\h2o\01_4150-4500_HITEMP2010.par')
wn_h2o, absorb_h2o = s_h2o.get("absorbance", wunit="cm-1", return_units=False)
wavelength_h2o = 1e7 / wn_h2o
line_h2o, = ax1.plot(wavelength_h2o, absorb_h2o, color=color_h2o, linewidth=2.0, label=f"HO ({temperature} K)", zorder=2)

# === POKAZATEL superlines ===
#pokazatel_data = np.loadtxt("filtr1e-24_4290(pm20)cm-1_POKAZATEL_1800K.txt")
pokazatel_path = f"data/4290(pm20)cm-1_POKAZATEL_{temperature}K.txt"
pokazatel_data = np.loadtxt(pokazatel_path)
wn_pok = pokazatel_data[:, 0]
intensity_pok = pokazatel_data[:, 1]
mask_pok = (wn_pok >= wmin) & (wn_pok <= wmax)
wn_pok_filtered = wn_pok[mask_pok]
intensity_pok_filtered = intensity_pok[mask_pok]
wavelength_pok = 1e7 / wn_pok_filtered
ax2.vlines(wavelength_pok, ymin=1e-25, ymax=intensity_pok_filtered,
           color='gray', linewidth=2.0, alpha=0.75, label="POKAZATEL superlines", zorder=1)

# === W2020 pechody ===
wn_w2020 = []
with open("data/4290(pm20)cm-1_ordered_transitions_H216O_W2020.txt", "r") as f:
    for line in f:
        parts = line.strip().split()
        try:
            wn = float(parts[0])
            if wmin <= wn <= wmax:
                wn_w2020.append(wn)
        except ValueError:
            continue
wavelength_w2020 = 1e7 / np.array(wn_w2020)
max_abs = max(absorb_co.max(), absorb_h2o.max())
ax1.vlines(wavelength_w2020, ymin=0, ymax=max_abs,
           color='lime', linestyle='dotted', linewidth=1.5, alpha=0.9, label="W2020 transitions", zorder=1)

# === Omez vku kivek kvli legend ===
ax1.set_ylim(0, 1.33 * max_abs)
ax2.set_ylim(0, 1.33 * np.max(intensity_pok_filtered))

# === Titulek grafu ===
#plot_title = f"Spectral survey at {temperature} K\n($L_{{opt}}=10$ cm, $p=1$ atm)  $X_{{CO}}={x_co}$, $X_{{H_2O}}={x_h2o}$"
#plt.title(plot_title, fontsize=14, loc='center', pad=30)

# === Popisy os ===
ax1.set_xlabel("Wavelength [nm]", fontsize=14)
ax1.set_ylabel("Absorbance (HITEMP)", fontsize=14, color='black')

ax2.set_ylabel("Line intensity (POKAZATEL)", fontsize=14, color='gray')
ax2.yaxis.offsetText.set_x(1.05)  # posune exponent vce doprava

ax1.tick_params(labelsize=12)
ax2.tick_params(labelsize=12)
ax1.grid(True, linestyle='--', alpha=0.5)

# === Horn osa s vlnovm slem ===
def wavelength_to_wavenumber(x): return 1e7 / x
def wavenumber_to_wavelength(x): return 1e7 / x
secax = ax1.secondary_xaxis('top', functions=(wavelength_to_wavenumber, wavenumber_to_wavelength))
secax.set_xlabel("Wavenumber [cm]", fontsize=14, labelpad=10)
secax.tick_params(labelsize=12)
secax.set_xlim(wmin, wmax)

# === Legenda nad grafem ===
legend_elements = [
    Line2D([0], [0], color=color_co, lw=2, label=f"CO (HITEMP {temperature} K)"),
    Line2D([0], [0], color=color_h2o, lw=2, label=f"HO (HITEMP {temperature} K)"),
    Line2D([0], [0], color='gray', lw=2, linestyle='solid', label=f"HO superlines (POKAZATEL {temperature} K)"),
    Line2D([0], [0], color='lime', lw=1.5, linestyle='dotted', label="MARVEL HO transitions (W2020)")
]

ax1.legend(handles=legend_elements,
           loc='upper center',
           bbox_to_anchor=(0.5, 0.99),
           ncol=2,
           fontsize=11,
           frameon=True,
           fancybox=True,
           borderpad=0.75)

# === Posun grafu dol kvli legend ===
plt.subplots_adjust(top=0.75)

plt.tight_layout()
plt.show()
```
